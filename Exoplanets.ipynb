```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
planets = pd.read_csv('all_exoplanets_2021.csv')
```

```python
planets
```

```python
# General Description
planets.describe()
```

```python
# How many new stars have been discovered?
planets['Num Stars'].sum()
```

```python
# How many new planets have been discovered?
planets['Num Planets'].sum()
```

```python
# The average mass of the planets?
planets['Mass'].mean()
```

```python
# Who is the biggest planet?
planets.loc[planets['Mass']==planets['Mass'].max()]
```

```python
planets['Mass'].max()
```

```python
# Who is the smallest planet?
planets.loc[planets['Mass']==planets['Mass'].min()]
```

```python
planets['Mass'].min()
```

```python
sns.lmplot(x='Num Stars', y='Num Planets', data=planets)
```

```python
# The most common number of planets in a system
planets['Num Planets'].mode()
```

```python
planets['Num Stars'].mode()
```

```python
# What is the number of planets approximately like earth (boundries choosen arbitraly)
planets.loc[(planets['Mass']>0.5) & (planets['Mass']<2), 'Num Planets'].sum()
```

```python
# Is there a trend of discovering more planets recently?
sns.countplot(x='Discovery Year', data=planets)
plt.xticks(rotation=90)
```

```python
planets_by_year=planets['Discovery Year'].value_counts().sort_index()
```

```python
planets_by_year.cumsum().plot(figsize = (11,6));
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Exoplanets')
plt.title('Number of Known Exoplanets')
```
